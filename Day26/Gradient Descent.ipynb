```python
from sklearn.datasets import make_regression
```

```python
import numpy as np
```

```python
X , y = make_regression(n_samples=4 , n_features=1, 
                        n_informative=1,n_targets=1 , noise=80,
                       random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
# m value
```

```python
lr.intercept_
# b value
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
```

```python
# m = 78.35
# b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X , lr.predict(X),color='r',label='OLS')
plt.plot(X,y_pred,color='b',label='b=0')
plt.legend()
```

```python
m = 78.35
b = 0

loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python

LR = 0.1
#LR * Slope --> Step Size

step_size = loss_slope * LR
step_size
```

```python
#new b
b = b - step_size
b
```
